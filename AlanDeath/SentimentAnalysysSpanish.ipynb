Salida y significado

La funci√≥n sentiment(text) devuelve un n√∫mero entre 0 y 1. Esta es la probabilidad de que la variable de cadena text sea "positiva". Las probabilidades bajas significan que el texto es negativo (n√∫meros cercanos a 0), las probabilidades altas (n√∫meros cercanos a 1) significan que el texto es positivo. El espacio entre ellos corresponde a textos neutrales.
